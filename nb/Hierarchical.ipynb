# Hierarchical Clustering

```python
import numpy as np
import glob

mp3s = glob.glob("npy/*.npy")
```

```python
pred_list = [np.load(f) for f in mp3s]
```

```python
for i, arr in enumerate(pred_list):
    if i==0:
        pred_arr = arr
    else:
        pred_arr = np.append(pred_arr,arr, axis=0)
```

```python
font = {'family' : 'arial',
        'weight' : 'normal',
        'size'   : 18}
import matplotlib
matplotlib.rc('font', **font)
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
%matplotlib inline

linked = linkage(pred_arr, 'ward', metric='euclidean')

labelList = [s.split('/')[-1] for s in mp3s]

plt.figure(figsize=(10, 7))
dendrogram(linked,
            orientation='left',
            labels=labelList,
            distance_sort='descending',
            leaf_font_size=10,
            show_leaf_counts=True)
plt.show()
```

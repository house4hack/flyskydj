# Mel Spectogram

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
filename = "output.wav"
```

```python
import IPython.display as ipd
ipd.Audio(filename)
```

```python
x, sr = librosa.load(filename)
```

```python
plt.figure(figsize=(14, 5))
librosa.display.waveplot(x[:1000], sr=sr)
```

```python

spect = librosa.feature.melspectrogram(y=x, sr=sr,n_fft=2048, hop_length=1024, n_mels=224,fmin=0)
spect = librosa.power_to_db(spect, ref=np.max)
fig = plt.figure(figsize=(15,10))
librosa.display.specshow(spect, x_axis="time", y_axis="mel")
```
